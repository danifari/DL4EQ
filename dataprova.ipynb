```python
from obspy import UTCDateTime
from obspy.clients.fdsn import Client
import matplotlib.pyplot as plt

# Initialize the client
client = Client("IRIS")

# Define the time window for the query
event_time = UTCDateTime("2016-08-24T00:00:00")  # Starting time (Amatrice Earthquake)
start_time = event_time  # Start time
end_time = event_time + 4 * 86400  # End time (4days)

# Define station details
station_code = "ANMO"  
network_code = "IU"    
location_code = "*"   
channel_code = "LHZ"  


st = client.get_waveforms(
    network=network_code, 
    station=station_code, 
    location=location_code, 
    channel=channel_code, 
    starttime=start_time, 
    endtime=end_time
)

st = client.get_waveforms(network_code, station_code, location_code, channel_code, start_time, end_time)

# Save the waveform data to your Downloads folder in WSL
output_file = "/mnt/c/Users/Utente/Downloads/prova.mseed"  # WSL path format

# Save the stream to the file
st.write(output_file, format="MSEED")

print(f"Waveform data saved to {output_file}")

```
